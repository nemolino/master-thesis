### Generation of a random application

```python
from random import random
from app_generator import generate_path_apps, generate_tree_apps, generate_graph_apps
```

```python
OUTPUT_FOLDER = "output/apps/"
```

```python
vehicle_min_cores, vehicle_max_cores = 5, 75
edge_min_cores, edge_max_cores = 80, 250
cloud_min_cores, cloud_max_cores = 350, 900
prob_vehicle_needs_camera = 0.5
prob_vehicle_needs_gpu = 0.35
prob_edge_needs_camera = 0.35
prob_edge_needs_gpu = 0.5
min_bandwidth, max_bandwidth = 50, 200
def get_latency():
    r = random()
    if r < 0.15:    return 2
    elif r < 0.6:   return 3
    else:           return 4

PARAMETERS = (
    vehicle_min_cores, vehicle_max_cores, edge_min_cores, edge_max_cores, cloud_min_cores, cloud_max_cores, 
    prob_vehicle_needs_camera, prob_vehicle_needs_gpu, prob_edge_needs_camera, prob_edge_needs_gpu, 
    min_bandwidth, max_bandwidth, get_latency
)
```

```python
# generazione dataset

for T in range(3,16):

    for suffix in ('a'):
        generate_path_apps(
            T=T, 
            prefix = OUTPUT_FOLDER,
            suffix = suffix,
            rp_count = 5,
            parameters = PARAMETERS
        )
    
    
    for suffix in ('a','b','c','d'):

        generate_tree_apps(
            T=T, 
            prefix = OUTPUT_FOLDER,
            suffix = suffix,
            rp_count = 5,
            parameters = PARAMETERS
        )
        
        generate_graph_apps(
            T=T, 
            prefix = OUTPUT_FOLDER,
            suffix = suffix,
            rp_count = 5,
            parameters = PARAMETERS
        )
    
```
